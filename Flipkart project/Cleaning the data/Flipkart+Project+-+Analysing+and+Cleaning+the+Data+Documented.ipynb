```python
def brandprediction(x):
    x = x.strip('[')
    x = x.strip(']')
    x = x.replace('"','')
    m = x.split(">>")
    for i in range(len(m)):
        m[i] = m[i].replace(' ','')
    for i in range(len(m)-2,0,-1):
        if m[i] in m[-1]:
            m[-1] = m[-1].replace(m[i],'')
    return m[-1]
```

```python
def commons(x):
    x = x.strip('[')
    x = x.strip(']')
    x = x.replace('"','')
    #x = x.replace('\'',' ')
    m = x.split(">>")
    
    return m
```

```python
def gettingbrand(x):
    n = commons(x)
    l = n[0:-1]
    xl = []
    for i in l:
        m = i.split()
        for j in m:
            xl.append(j)
    y = dfbrandnan['product_name'][dfbrandnan]
    ylist = y.split()
    brand = ''
    for i in range(len(ylist)):
        if y[i] not in xl:
            brand = brand + y[i]
    return brand    
    
```

```python
brand_list = []
for row_index,row in dfbrandnan.iterrows():
    x = row['product_category_tree']
    n = commons(x)
    y = dfbrandnan['product_name'][row_index]
    ylist = y.split()
    for j in range(len(n)):
        if n[j] in ylist:
            l = n[0:-1]
        else:
            l = n
    xl = []
    for i in l:
        m = i.split()
        for j in m:
            xl.append(j)
    #y = dfbrandnan['product_name'][row_index]
    #ylist = y.split()
    #print(ylist)
    brand = ''
    #brand_list = []
    for i in range(len(ylist)):
        #print(ylist[i])
        if ylist[i] not in xl:
            brand = brand + ' ' + ylist[i]
    brand = brand.strip()
    brand_list.append(brand)
   
```
