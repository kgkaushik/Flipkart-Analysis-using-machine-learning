```python
value_total_list = []
```

```python
def getvalues(x):
    values = []
    value_count = x.count('value')
    for i in range(value_count):
        m = x.index('value')
        x = x[m+9:]
        m = x.index('"')
        y = x[:m]
        x = x[m:]
        values.append(y)
    return values
```

```python
for row_index,row in df_clothing.iterrows():
    value_list = []
    try:
        value_list = getvalues(row['product_specs_refined'])
    except ValueError:
        value_list = 'No Values'
    value_total_list.append(value_list)
```

```python
value_total_list
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(value_total_list).transform(value_total_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.6, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
df_cleared = pd.read_csv('flipkart_final_cleared.csv')
df_cleared.head()
```

```python
df_clothing_with_rating = df_cleared[df_cleared['product_category']=='Clothing ']
```

```python
value_list_with_rating = []
for row_index,row in df_clothing_with_rating.iterrows():
    value_list = []
    try:
        value_list = getvalues(row['product_specs_refined'])
    except ValueError:
        value_list = ['No Values']
    value_list.append(str(int(row['product_rating'])))
    value_list_with_rating.append(value_list)
```

```python
value_list_with_rating[-4]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(value_list_with_rating).transform(value_list_with_rating)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.4)
```

```python
apriori(df, min_support=0.4, use_colnames=True)
```

```python
apriori(df, min_support=0.4, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.4) ]
```

# This shows that the casual pack of 1 selling is the most frequent choice of flipkart

```python
itemsets = frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.1) ]
```

```python
type(itemsets['itemsets'])
```

```python
def to_string(x):
    return str(x)
```

```python
itemsets['itemsets'].apply(to_string)
```

```python
itemsets.info()
```

```python
rule_list_total = []
for row_index,row in itemsets.iterrows():
    rule_list = []
    if row['itemsets'][0] in ['0','1','2','3','4','5']:
        rule_list.append(str(row['itemsets']))
        rule_list.append(row['support'])
        rule_list_total.append(rule_list)
```

```python
rule_list_total
```

```python
rule_df_clothing = pd.DataFrame(data=rule_list_total,columns=['rating_specificationvalue','support'])
```
